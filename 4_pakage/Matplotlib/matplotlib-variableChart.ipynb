## Bar plot

### 종류
- bar
- barh

```python
y = np.random.randint(0, 5, 10)
x = range(10)
error = np.random.randint(1, 5, 10)

plt.bar(x, y, yerr = error, align = 'center', alpha = 0.5)
```

```python
plt.barh(x, y, xerr = error, align = 'center', alpha = 0.6)
```

```python
y1 = np.random.randint(5, 10, 5)
y2 = np.random.randint(1, 5, 5)
x = range(5)

plt.bar(x, y1)
plt.bar(x, y2)
plt.tight_layout()
```

```python
y1 = np.random.randint(1, 5, 5)
y2 = np.random.randint(1, 5, 5)
x = np.arange(5)

plt.bar(x, y1, 0.4, alpha = 0.6, align = 'edge', color = 'b', label = '1')
plt.bar(x + 0.4, y2, 0.4, alpha = 0.6, align = 'edge', color = 'r', label = '2')

plt.title("Random")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc = 0)
plt.xticks(x + 0.4, (list('ABCDE')))
```

## Pie plot

```python
data = np.random.randint(1, 6, 5)
labels = list("ABCDE")
colors = ['r', 'g', 'b', 'w', 'k']

plt.pie(data, labels = labels, colors = colors, autopct = '%1.1f%%')
```

## Scatter plot

```python
x = np.random.randint(1, 500, 500)
y = np.random.randint(1, 500, 500)

plt.scatter(x, y)
```

```python
x = np.random.randint(1, 10, 10)
y = np.random.randint(1, 10, 10)
colors = np.random.rand(10)
area = np.pi * np.arange(10, 30, 2) ** 2

plt.scatter(x, y, s = area, c = colors, alpha = 0.6)
```
